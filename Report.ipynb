## 2. Problem definition (<300 words)

The aim of this project is to be able to build a new GSM (genome scale model) from scratch, to refine it and to finally validate it using existing data from experiments from other studies.

Phaeobacter inhibens is a marine heterotrophic bacterium, which is  metabolically very versatile, using a lot of different carbon sources (amino acids, carbohydrates and simple organic acids).

Bacteria belonging to Roseobacter-clade represent around the 25% of marine microbial communities in coastal regions and polar oceans. They have adapted to disparate ecological niches and they are characterized by a wide range of physicological characteristics such as the capacity of forming biofilms, aerobic photosynthesis or Degradation of dimethylsulfoniopropionate (DMSP).In the case of P.inhibens DSM17395 it can synthesize secondary metabolites that have antibiotic and probiotic effects.

In this project, there is also the implementation of the production of TDA (tropodithietic acid), an antibacterial compound that could have an interesting impact on the market since it has been proven to have a very large application potential.

TDA has antibacterial action against bacteria outside the marine or aquatic environment, including human-pathogenic bacteria which contains both Gram-negative and Gram-positive bacteria, like Escherichia coli, Pseudomonas aeruginosa. Salmonella enterica serovar Typhimurium SL1344 and Staphylococcus aureus NCTC 12493.

## 3. Reconstruction of a new GSM for your cell factory host of interest (<1500 words)

In order to reconstruct the new GSM for P.inhibens, carve me was used and gapfilled with LB. The refseq for P.inhibens DSM 17395 is "GCF_000154765.2". 
Unfortunately the carveme left some issues on the SBML model obtained, it was missing the boundaryCondition. This was solved by adding it manually using a loop. Afterwards a new SBML model was obtained, this time being valid.

From it, a SBML model validation was run (with a positive outcome) and then a memote validation was run, giving very good MEMOTE results as it is shown on the picture below. The average memote grade 86% approved.

Due to this results, it was assessed that there was no need to make any changes to it to refine it.

```python
Image("MEMOTE_RESULTS_GRAPHIC.png")
```

After getting the model, a reconstruction and high analysis of its stiochiometry matrix (S) was conducted. By doing so, important information such as the degrees of freedom (893) and the relations ship between the metaoblites and reactions involved in the model are obtained. On the picture below it is shown the visualization of the stoichiometry matrix.

```python
Image("Stoichiometry_matrix_S.png")
```

Following up with the stoichiometry matrix, the top 10 most connected metabolites -the ones involved in the highest number of reactions- were found using the model constructed. The top 10 metabolites are:
 - H+ (h_c)
 - H2O (h2o_c)
 - ATP (atp_c)
 - Phosphate (pi_c)
 - ADP (adp_c)
 - Coenzyme A (coa_c)
 - Nicotinamide adenine dinucleotide (nad_c)
 - Nicotinamide adenine dinucleotide - reduced (nadh_c)
 - Diphosphate (ppi_c)
 - H+ (h_p)
